## Adding Layers

A neural network becomes deep when we have 2 or more hidden layers. An hidden layer has values that we do not deal with, i.e. it found between the input and output layers.  
Each hidden layer has a set of neurons, each of which is associated with its weight set (one weight per input feature). Thus, for a 3-neurons layer for 4-feature inputs, we would have a (3,4) weight matrix.  
To add a layer, we want to use the previous layer's output as input to the new layer. This means that the shape of the size of this layer should match the previous layer's output. The new layer can have as many neurons as we want, but each neuron has to have a number of weights = the number of neurons of the previous layer (since this will be the new input sets).

```python
inputs = [
    [1,2,3,2.5],
    [2., 5.,-1, 2],
    [-1.5, 2.7, 3.3, -0.8]
]

weights_1 = [
    [0.2, 0.8, -0.5, 1],
    [0.5, -0.91, 0.26, -0.5],
    [-0.26, -0.27, 0.17, 0.87]
]
biases_1 = [2,3, 0.5]

weights_2 = [
    [0.1, -0.14, 0.5],
    [-0.5, 0.12, -0.33],
    [-0.44, 0.73, -0.13]
]
biases_2 = [-1, 2, -0.5]
```

```python
layer_1_outputs = np.dot(inputs, np.array(weights_1).T) + biases_1
layer_2_outputs = np.dot(layer_1_outputs, np.array(weights_2).T) + biases_2
```

## Dense Layer Class

So far we have used what's called a **dense** (i.e., fully connected) layer. We can define a class for this kind of layer, where:  
1. **weights** are initialized randomly (we could have other options, in particular in case we want to fine tune a pretrained models we would load pretrained weights)
2. **forward pass** is a method that lets data pass from beginning to end. We have many options (e.g., with looping data around), but in this case we implement a regular forward pass

```python
class dense_layer:
    def __init__ (self, n_inputs, n_neurons):
        #weights and biases are initalized as transposed already
        self.weights = 0.01*np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    def forward(self, inputs):
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
X, y = spiral_data(samples=100, classes=3)
dense1 = dense_layer(2, 3)
dense1.forward(X)

print(dense1.output[:5])
```
